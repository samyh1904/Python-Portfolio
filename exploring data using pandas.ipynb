```python
#looking at correlations between columns
corr = df.corr(numeric_only = True)
corr
```

```python
df.columns
```

```python
#generating a heatmap where the lighter colour is highly correlated
sns.heatmap(corr, annot = True)

plt.rcParams['figure.figsize'] = (20,15)

plt.show()
```

```python
#grouping data
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only = True).sort_values(by='2022 Population', ascending = False)
df2
```

```python
#transposed the data for the following figure
df3 = df2.transpose()
df3
```

```python
# visualising data
df3.plot(figsize = (20,8))
```
